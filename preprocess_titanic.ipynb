```python
def f(name) :
    if '.' in name:
        return  name.split(',')[1].split('.')[0].strip()
    else:
        return 'Uknown'
```

```python
df['Name'].apply(f)
```

```python
for i in df['Name']:
  print(i.split(',')[1].split('.')[0])
```

```python
def g(tt):
    if tt in ['Mr']:
        return 1
    elif tt in ['Master']:
        return 3 
    else :
        return 2      
```

```python
df['title'] = df['Name'].apply(f).apply(g)
df
```

```python
df.drop(['PassengerId','Ticket','Name'],axis=1)
```

```python
t = pd.crosstab(df['title'],df['Survived'])
t
```

```python
t_pct = t.div(t.sum(1).astype(float),axis=0)
t_pct
```

```python
t_pct.plot.bar(stacked=True)
```
